```python
# Import Modules from the Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

### Let's create some data

```python
# Seed of the Random Number Generator
np.random.seed(1001)
```

```python
# Data Generating Function
def f(x, sigma):
    return 0.3*x - 0.73 + np.random.randn(x.shape[-1])*sigma
```

```python
# Training Data
N_data = 100
x_data = np.random.rand(N_data)*10
y_data = f(x_data, 0.2)
```

```python
plt.plot(x_data, y_data, 'o')
```

### Now, let's sample some random model parameters from a normal distribution

```python
N_model = 50

W = np.random.randn(2, N_model)
W[0, :] = W[0, :]* 0.1 # linear term
W[1, :] = W[1, :]* 0.5   # constant term
```

```python
W[0].std(), W[1].std(), W[0].mean(), W[1].mean()
```

Hint:

$x \sim \mathcal{N}(\mu, \sigma^2)$

is equivalent to

### Calculate the y-values...

```python
x = np.linspace(0, 10, 101)
X = np.vander(x, 2)
```

```python
W.shape, X.shape
```

```python
y = X @ W
```

```python
y.shape
```

### ... and visualize the result:

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[1].plot([0], [0])

for i in range(N_model):
    ax[0].plot(x, y[:,i], label=f"{i}")
    ax[1].scatter(W[0,i], W[1,i], marker="x", label=f"{i}")

ax[0].plot(x_data, y_data, 'o', color="blue")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[1].set_xlabel("W0")
ax[1].set_ylabel("W1")

if N_model <= 10:
    ax[0].legend()
    ax[1].legend()
```

### Let's assume a simple error model for our data:

\begin{equation}
t = y(x,w) + \epsilon
\end{equation}

\begin{equation}
\epsilon \sim \mathcal{N}(0, s^2 \mathbb{I}), \qquad s^2 \mathbb{I} = S
\end{equation}




\begin{equation}
p(t | x, w, S) = \mathcal{N}(t| y(x,w), S)
\end{equation}

```python
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
```

```python
p_choice = []
w_choice = [0,0]
delta_jihe = []
for j in range(N_model):
    p1 = 1
    w0 = W[0,j]
    w1 = W[1,j]
    y0 = x_data*w1+w0
    delta = 0

    for i in range(100):
        delta = delta + (y_data[i] - y0[i])**2
        p1 = p1 * normal_dist(y_data[i],y0[i],0.2) 
    delta_jihe.append(delta)
    p_choice.append(p1)
maxdelta = min(delta_jihe)
indexx = delta_jihe.index(maxdelta)
```
